```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import preprocessing as per
```

```python
#importing datasets
data1= pd.read_csv('All Dataset.csv')
```

```python
data1
```

```python
merged=data1
```

```python
dataDrop = merged.drop(columns=['id','num',])
```

```python
dataDrop
```

```python
dropData = merged['num']
```

```python
dropData
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

```python
# Begin by importing all necessary libraries
import pandas as pd
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dataDrop,dropData, test_size=0.20, random_state=27)
```

```python
print(X_train)
print(y_train)
```

```python
from sklearn import svm
```

```python
SVC_model = svm.SVC()
# KNN model requires you to specify n_neighbors,
# the number of points the classifier will look at to determine what class a new point belongs to
#KNN_model = KNeighborsClassifier(n_neighbors=5)
```

```python
SVC_model.fit(X_train, y_train)
```

```python
SVC_prediction = SVC_model.predict(X_test)
```

```python
# Accuracy score is the simplest way to evaluate
print(accuracy_score(SVC_prediction, y_test))
# But Confusion Matrix and Classification Report give more details about performance
print(confusion_matrix(SVC_prediction, y_test))
print(classification_report(SVC_prediction, y_test))
```

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from xgboost.sklearn import XGBClassifier
```

```python
xgb_model = XGBClassifier()
```

```python
xgb_model.fit(X_train, y_train)
```

```python
xgb_prediction = xgb_model.predict(X_test)
```

```python
# Accuracy score is the simplest way to evaluate
print(accuracy_score(xgb_prediction, y_test))

# But Confusion Matrix and Classification Report give more details about performance
print(confusion_matrix(xgb_prediction, y_test))
print(classification_report(xgb_prediction, y_test))
```
